```python
import numpy as np
import pandas as pd
import random
```

## Mage 1

```python
#creating dice rols to represent mage health and mage damage in the next few cells
#1,000,000 trials
#Mage 1 has 6 dice that are 4 sided
n_trials = nrows = 1_000_000
n_dice = ncols = 6
mage1_dmg_roll = np.random.choice([2,3,4,5], n_trials * n_dice).reshape(nrows, ncols)
mage1_dmg_roll = pd.DataFrame(mage1_dmg_roll, columns={1,2,3,4,5,6})
mage1_dmg_roll['mage1_dmg_sum'] = mage1_dmg_roll.sum(axis=1)
mage1_dmg_roll
```

```python
n_trials = nrows = 1_000_000
n_dice = ncols = 6
mage1_health_roll = np.random.choice([2,3,4,5], n_trials * n_dice).reshape(nrows, ncols)
mage1_health_roll = pd.DataFrame(mage1_health_roll, columns={1,2,3,4,5,6})
mage1_health_roll['mage1_health_sum'] = mage1_health_roll.sum(axis=1)
mage1_health_roll
```

## Mage 2

```python
#Mage 2 has 4 dice of 6 sides each
n_trials = nrows = 1_000_000
n_dice = ncols = 6
mage2_dmg_roll = np.random.choice([2,3,4,5], n_trials * n_dice).reshape(nrows, ncols)
mage2_dmg_roll = pd.DataFrame(mage2_dmg_roll, columns={1,2,3,4,5,6})
mage2_dmg_roll['mage2_dmg_sum'] = mage2_dmg_roll.sum(axis=1)
mage2_dmg_roll
```

```python
n_trials = nrows = 1_000_000
n_dice = ncols = 6
mage2_health_roll = np.random.choice([2,3,4,5], n_trials * n_dice).reshape(nrows, ncols)
mage2_health_roll = pd.DataFrame(mage2_health_roll, columns={1,2,3,4,5,6})
mage2_health_roll['mage2_health_sum'] = mage2_health_roll.sum(axis=1)
mage2_health_roll
```

```python
#create a DF with all the pertinent info: health and damage for respective mages
mage_duel = pd.DataFrame({"mage1_health" : mage1_health_roll.mage1_health_sum, 
                          "mage1_dmg" : mage1_dmg_roll.mage1_dmg_sum,
                          "mage2_health" : mage2_health_roll.mage2_health_sum,
                         "mage2_dmg" : mage2_dmg_roll.mage2_dmg_sum})
mage_duel
```

```python
#Add columns to tally results: health minus damage
#Add a column to return the winner
mage_duel['mage1_result'] = mage_duel.mage1_health - mage_duel.mage2_dmg
mage_duel['mage2_result'] = mage_duel.mage2_health - mage_duel.mage1_dmg
mage_duel['Mage1_Win'] = mage_duel.mage1_result > mage_duel.mage2_result

mage_duel
```

```python
#Give us the average of mage 1 wins.  Only 46%! 
#This result is contrary to my hypothesis that mage 1 was slightly more powerful
mage_duel.Mage1_Win.mean()
```
